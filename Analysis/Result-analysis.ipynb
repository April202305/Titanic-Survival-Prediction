```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Organize data
data = {
    'Decision Tree': {
        'Accuracy': 0.8045,
        'Precision (Class 0)': 0.80,
        'Recall (Class 0)': 0.90,
        'F1 (Class 0)': 0.85,
        'Precision (Class 1)': 0.80,
        'Recall (Class 1)': 0.65,
        'F1 (Class 1)': 0.72,
        'Macro Avg F1': 0.78,
        'Weighted Avg F1': 0.80
    },
    'Logistic Regression': {
        'Accuracy': 0.8101,
        'Precision (Class 0)': 0.86,
        'Recall (Class 0)': 0.83,
        'F1 (Class 0)': 0.84,
        'Precision (Class 1)': 0.74,
        'Recall (Class 1)': 0.78,
        'F1 (Class 1)': 0.76,
        'Macro Avg F1': 0.80,
        'Weighted Avg F1': 0.81
    }
}

df = pd.DataFrame(data).T

# Draw a comparison table
styled_df = df.style.format("{:.2f}").highlight_max(color='lightgreen', axis=0)
styled_df.set_caption("Model Performance Comparison").set_table_styles([{
    'selector': 'caption',
    'props': [('font-size', '16px'), ('font-weight', 'bold')]
}])

# Draw a column comparison chart
plt.figure(figsize=(14, 8))

# Select key indicators for visualization
selected_metrics = [
    'Accuracy', 
    'Precision (Class 0)', 
    'Recall (Class 0)', 
    'F1 (Class 0)',
    'Precision (Class 1)', 
    'Recall (Class 1)', 
    'F1 (Class 1)',
    'Macro Avg F1', 
    'Weighted Avg F1'
]

# Adjust subgraph layout
ax = df[selected_metrics].plot(kind='bar', figsize=(14, 8), width=0.8, rot=0)

# Chart beautification settings
plt.title('Model Performance Comparison', fontsize=16, pad=20)
plt.ylabel('Score', fontsize=12)
plt.xlabel('Model', fontsize=12)
plt.ylim(0, 1.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numerical labels
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2.,
            height + 0.02,
            f'{height:.2f}',
            ha='center',
            va='bottom',
            fontsize=10)

# Adjust the position of the legend
plt.legend(
    title='Metrics',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    borderaxespad=0.,
    fontsize=10
)

# Display chart
plt.tight_layout()
plt.show()

# Display Table
styled_df
```
